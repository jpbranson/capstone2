```python
def getCityIndex(city):
        matches = []
 
        endpoint_news ="https://www.numbeo.com/api/indices?api_key=37781hcsp9me97&query="+city

        resp = requests.get(endpoint_news   )
        
        data =  resp.json()

        print(  retrieveField(data,"health_care_index") 
              +"\t" + retrieveField(data,"crime_index") 
              +"\t" +retrieveField(data,"purchasing_power_incl_rent_index")
              +"\t" +retrieveField(data,"cpi_index") 
              +"\t" +retrieveField(data,"pollution_index") 
              +"\t" +retrieveField(data,"traffic_index") 
              +"\t" +retrieveField(data,"quality_of_life_index") 
              +"\t" +retrieveField(data,"rent_index")
              +"\t" +retrieveField(data,"property_price_to_income_ratio") 
              +"\t" +retrieveField(data,"traffic_co2_index")
              +"\t" +str(data["name"]) 
            )   
 
```

```python

def retrieveField(data,fieldName):         
    fieldValue = ""
    try:
        fieldValue= str(round(data[fieldName],2))
    except:
        fieldValue="NA"  

    return fieldValue
```
